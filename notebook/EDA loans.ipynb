```python
#Load packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("/Users/aidanastridge/Documents/LoanData.csv")
```

```python
#Function for explatory data analysis
def eda(x):
    print(x.name)
    print("Mean:", np.mean(x))
    print("Standard Deviation:", np.std(x))
    print("Variance:",np.var(x))
    print("Minimum:",np.min(x))
    print("Q25:", np.quantile(x,.25))
    print("Median:",np.quantile(x,.5))
    print("Q75:",np.quantile(x,.75))
    print("Maximum:",np.max(x))
    print("")

#Function for dispersion
def disp(x):
    print(x.name)
    print("Range:", np.max(x)-np.min(x))
    q3,q1 = np.percentile(x, [75,25])
    print("IQR:", q3-q1)
    print("CV:", np.std(x)/np.mean(x))
    print("")
```

```python
#Eda loan amount
eda(round(df["LoanAmount"]))
```

```python
#Loan amount by commercial range
disp(round(df.loc[df['BusinessLine'] == 'Commercial', 'LoanAmount']))
```

```python
#Loan amount by corporate range
disp(df.loc[df['BusinessLine'] == 'Corporate', 'LoanAmount'])
```

```python
#Loan amount exploratory data analysis by provice (ON, BC, AB)
for x in list(df['Region'].unique()):
    print(x)
    eda(df.loc[df['Region'] == x, 'LoanAmount']) 
```
